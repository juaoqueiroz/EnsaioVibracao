```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import openpyxl as opp
```

```python
vibration_data = pd.read_csv("../Datasets/dataset_00.csv", sep=',')
```

```python
# Configurar o tamanho da imagem (largura x altura)
plt.figure(figsize=(20, 8))

plt.plot(vibration_data['time'], vibration_data['aceleration'])

plt.xlabel('Tempo(s)', fontsize = 19)
plt.ylabel('Acelerao(g)', fontsize = 19)

# Adicionar um ttulo ao grfico
plt.title('Acelerao X Tempo', fontsize = 19)


# Exibir o grfico
plt.show()
```

```python
# Scatter plot com cores baseadas na coluna 'rpm'
plt.figure(figsize=(20, 8))
plt.scatter(vibration_data['time'], vibration_data['aceleration'], c=vibration_data['rpm'], cmap='viridis')

plt.xlabel('Tempo(s)', fontsize=19)
plt.ylabel('Acelerao(g)', fontsize=19)
plt.title('Acelerao X Tempo (Colorido por RPM)', fontsize=19)

# Adicionar barra de cores
cbar = plt.colorbar()
cbar.set_label('RPM', fontsize=16)

# Exibir o grfico
plt.show()
```
